```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL
url = 'http://www.dream-pro.info/~lavalse/LR2IR/search.cgi?mode=ranking&courseid=13136'

# ページを取得してエンコーディングを指定してパース
res = requests.get(url)
res.encoding = 'cp932'
soup = BeautifulSoup(res.text, 'html.parser')

# 対象テーブル（4番目）を取得
tables = soup.find_all('table')
target_table = tables[3]

# pandas用にDataFrameを取得
df = pd.read_html(str(target_table))[0]

# BeautifulSoupでプレイヤー名とリンクを抽出
player_links = []
for row in target_table.find_all('tr')[1:]:  # ヘッダーを除く
    cols = row.find_all('td')
    if len(cols) >= 2:
        a_tag = cols[1].find('a')
        if a_tag and a_tag.get('href'):
            player_links.append(a_tag['href'].split("playerid=")[1])
        else:
            player_links.append(None)
    else:
        player_links.append(None)

# DataFrameにリンク列を追加
df['LR2ID'] = player_links

df = df.dropna(subset=[df.columns[3]])[['順位', 'LR2ID', 'スコア', 'PG', 'GR']]
```
